# Data Exploration                           

# Write a python program to find out top 15 jobs. Make a bar graph of salary as well. 

```python
# Grouping job_title, calculating mean of salary, and sorting it in descending order, and then selecting the top 15 jobs
top_jobs = data_frame['job_title'].value_counts().head(15).index
fifteen_job_title = data_frame[data_frame['job_title'].isin(top_jobs)]
mean_salary = fifteen_job_title.groupby('job_title')['salary'].mean().reset_index()
job_sorted = mean_salary.sort_values(by='salary', ascending=False)

# Create a figure with a specific size
plt.figure(figsize=(12, 8))

# Plotting a bar chart for the top 15 job titles and their average salaries
plt.bar(job_sorted['job_title'], job_sorted['salary'], width=0.4, color='skyblue')

# Setting labels and title for the plot
plt.xlabel('Job Title')
plt.ylabel('Salary')
plt.title('Top 15 Jobs')

# Rotating the x-axis labels for better visibility
plt.xticks(rotation=90, ha='right')

# Adjusting layout and displaying the plot
plt.tight_layout()
plt.show()
```

# Which job has the highest salaries? Illustrate with bar graph.

```python
# Calculating the highest salaries among different job titles in the data frame.
highest_salaries = data_frame.groupby('job_title')['salary'].sum()
highest_salaries.max()
```

```python
plt.figure(figsize=(30, 12))
# Create a bar chart with job titles on the x-axis and corresponding salaries on the y-axis
plt.bar(highest_salaries.index, highest_salaries.values, width=1, color='skyblue')

# Seting labels for x and y axes
plt.xlabel('Job Title')
plt.ylabel('Salary')

# Set the title of the plot
plt.title('Highest Salaries by Job Title')

# Rotate x-axis labels 
plt.xticks(rotation=90, ha='right')

# Adjust subplot parameters for a tight layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Write a python program to find out salaries based on experience level. Illustrate it through bar graph.

```python
# Calculate the total salary for each experience level by grouping the data frame
Salary_experience = data_frame.groupby('experience_level')['salary'].sum()
Salary_experience 
```

```python
plt.figure(figsize=(8, 5))
# Create a bar chart with experience levels on the x-axis and corresponding salaries on the y-axis
plt.bar(Salary_experience.index, Salary_experience.values, color='skyblue', width=0.4)

# Set labels for x and y axes
plt.xlabel('Experience Level')
plt.ylabel('Salary')

# Set the title of the plot
plt.title('Salary Based on Experience Level')

# Add a horizontal grid with dashed lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adjust subplot parameters for a tight layout
plt.tight_layout()

# Display the plot
plt.show()
```

# Write a Python program to show histogram and box plot of any chosen different variables. Use proper labels in the graph.

```python
import matplotlib.pyplot as plt

# Set the figure size for the plots
plt.figure(figsize=(15, 6))

# Create a subplot for the histogram
plt.subplot(1, 2, 1)
# Plot the histogram of the 'salary_in_usd' column from the data frame
data_frame['salary_in_usd'].hist()
# Set the title and axis labels for the histogram
plt.title('Histogram of Salary (USD)')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')

# Create a subplot for the box plot
plt.subplot(1, 2, 2)
# Plot the box plot of the 'salary_in_usd' column from the data frame
data_frame.boxplot(column='salary_in_usd')
# Set the title and axis labels for the box plot
plt.title('Salary Box Plot')
plt.ylabel('Salary (USD)')

# Display the plots
plt.show()
```
